##CASO ESTRATEGIA MOMENTUM
En este trabajo se va a realizar una simulación de estrategia de inversión de factor Momentum basada en el índice tecnológico estadounidense NASDAQ100.

Esta simulación se desarrollará en un horizonte temporal de 10 años, desde el 01/01/2014 hasta el 01/01/2024. Debido a que no todas las empresas del NASDAQ100 tienen datos históricos de los 10 años estudiados, se han utilizado únicamente las 82 empresas dentro del índice que sí cuentan con estos datos.

Debido a esta limitación, se ha decidido añadir como Bechmark, además del propio NASDAQ100, la cartera formada por los 82 valores con los que se trabajarán. Al haber permanecido estas empresas en el índice cabría esperar que estas hayan tenido un mejor rendimiento durante el periodo (survivor bias), por lo que sería justo tomarlo como punto de comparación de la estrategia que se va a desarrollar.

Se van a desarrollar dos variantes de esta estrategia Momentum:

  1)	Estrategia en la que cada mes (rebalanceo mensual) se comprarán las empresas que mejor lo hayan hecho en el periodo de observación y se venderán las que peor lo hayan hecho en ese mismo periodo. Se trata de una estrategia con posiciones largas y cortas. MOMENTUM LONG/SHORT

  2)	Estrategia en la que cada mes (rebalanceo mensual) se comprarán las empresas que mejor lo hayan hecho en el periodo de observación. En este caso únicamente existirán posiciones largas. MOMENTUM LONG

Para ambas estrategias se utilizarán 2 periodos de observación de rendimientos distintos, de forma que se puedan comparar distintos escenarios. Así, por una parte, se observarán los rendimientos del mes anterior para hacer los rebalanceos cada mes. Por otra parte, se observarán los rendimientos de los 6 meses anteriores, sin tener en cuenta el mes inmediatamente anterior, para evitar el efecto short-term reversal.

Este código define una función llamada assetPriceReg que realiza un análisis de regresión financiera. Primero, importa bibliotecas necesarias para el análisis de datos y regresión. Luego, utiliza pandas_datareader para obtener datos diarios de factores Fama-French de 5 factores. Ajusta y renombra estos factores, y los combina con los datos de las acciones que se han elegido.

La función calcula la prima de riesgo restando la tasa libre de riesgo a los rendimientos de las acciones. Posteriormente, ajusta tres modelos de regresión: CAPM, FF3 y FF5. Extrae estadísticas y coeficientes de estas regresiones y los organiza en un DataFrame.

Finalmente, imprime un resumen de los resultados de las regresiones utilizando la función summary_col. La función devuelve un DataFrame que contiene los coeficientes y estadísticas de los modelos.

```python
def assetPriceReg(df_stk):
    import pandas_datareader.data as web  # module for reading datasets directly from the web

    # Reading in factor data
    df_factors = web.DataReader('F-F_Research_Data_5_Factors_2x3_daily', 'famafrench')[0]
    df_factors.rename(columns={'Mkt-RF': 'MKT'}, inplace=True)
    df_factors['MKT'] = df_factors['MKT']/100
    df_factors['SMB'] = df_factors['SMB']/100
    df_factors['HML'] = df_factors['HML']/100
    df_factors['RMW'] = df_factors['RMW']/100
    df_factors['CMA'] = df_factors['CMA']/100

    df_stock_factor = pd.merge(df_stk,df_factors,left_index=True,right_index=True) # Merging the stock and factor returns dataframes together
    df_stock_factor['XsRet'] = df_stock_factor['Returns'] - df_stock_factor['RF'] # Calculating excess returns

    # Running CAPM, FF3, and FF5 models.
    CAPM = sm.ols(formula = 'XsRet ~ MKT', data=df_stock_factor).fit(cov_type='HAC',cov_kwds={'maxlags':1})
    FF3 = sm.ols( formula = 'XsRet ~ MKT + SMB + HML', data=df_stock_factor).fit(cov_type='HAC',cov_kwds={'maxlags':1})
    FF5 = sm.ols( formula = 'XsRet ~ MKT + SMB + HML + RMW + CMA', data=df_stock_factor).fit(cov_type='HAC',cov_kwds={'maxlags':1})

    CAPMtstat = CAPM.tvalues
    FF3tstat = FF3.tvalues
    FF5tstat = FF5.tvalues

    CAPMcoeff = CAPM.params
    FF3coeff = FF3.params
    FF5coeff = FF5.params

    # DataFrame with coefficients and t-stats
    results_df = pd.DataFrame({'CAPMcoeff':CAPMcoeff,'CAPMtstat':CAPMtstat,
                               'FF3coeff':FF3coeff, 'FF3tstat':FF3tstat,
                               'FF5coeff':FF5coeff, 'FF5tstat':FF5tstat},
    index = ['Intercept', 'MKT', 'SMB', 'HML', 'RMW', 'CMA'])


    dfoutput = summary_col([CAPM,FF3, FF5],stars=True,float_format='%0.4f',
                  model_names=['CAPM','FF3','FF5'],
                  info_dict={'N':lambda x: "{0:d}".format(int(x.nobs)),
                             'Adjusted R2':lambda x: "{:.4f}".format(x.rsquared_adj)},
                             regressor_order = ['Intercept', 'MKT', 'SMB', 'HML', 'RMW', 'CMA'])

    print(dfoutput)

    return results_df
```

Primero se recogen en la variable "ret" los precios de cierre ajustados de las acciones seleccionadas, para posteriormente calcular los retornos utilizando el logarítmo natural para obtener rendimientos logarítmicos mensuales. Por último muestra los resultados en una tabla.

#MOMENTUM LONG/SHORT
Primeramente se realizará la estrategia de Momentum Long/Short, en la que cada mes se compra el mejor decil y se vende el peor decil.

Para ello es necesario calcular primero dichos deciles, de forma que se obtengan las empresas que se comprarán y venderán cada mes. Cada decil estará compuesto de 8 empresas.

Para ello se asigna a las 8 mejores el número "1" (Las empresas que se compran), y a las 8 peores el "-1" (Las empresas que se venden).

Esto se hace tanto para el caso donde se observan los retornos del mes anterior como para el caso de los seis meses anteriores.

Aqui lo que se ha hecho es definir lo que serán los deciles que se compraran y venderan con la estrategia momentum.

Donde, z serán las 8 mejores acciones y i serán las 8 con peores rendimientos.

```python
i = int(len*0.1)
z=len-i
z=z/len
i=i/len
```

```python
decile_signal_1m = momentum_signal_1m.copy()
for col in decile_signal_1m.columns:
  decile_signal_1m[col] = np.where((decile_signal_1m[col]/len) > z, 1, np.where((decile_signal_1m[col]/len) <= i, -1, 0))
decile_signal_1m.replace(0, np.nan, inplace=True)
decile_signal_1m.head()
```

```python
decile_signal_6m = momentum_signal_6m.copy()
for col in decile_signal_6m.columns:
  decile_signal_6m[col] = np.where((decile_signal_6m[col]/len) > z, 1, np.where((decile_signal_6m[col]/len) <= i, -1, 0))
decile_signal_6m.replace(0, np.nan, inplace=True)
decile_signal_6m.head()
```

A continuación se sustituyen los 1 y -1, por los rendimientos mensuales y se agregan los retornos de cada mes conseguidos por cada una de las empresas de la estrategia.

```python
ret_df = ret_df.drop(ret_df.index[0:7])
```
